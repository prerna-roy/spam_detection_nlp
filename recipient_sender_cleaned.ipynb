```python
import pandas as pd 
import regex as re
```

```python
spam_email = pd.read_csv("/Users/neiltejnani/Downloads/emailspam_clean.csv")
spam_email
```

```python
def split_it(text):
    return re.findall(r'[\w\.-]+@[\w\.-]+\.\w+', text)

spam_email['Recipient']=spam_email['email_to'].apply(split_it)
spam_email['Recipient'] = spam_email['Recipient'].map(lambda x: str(x)[2:-2])

spam_email['Sender']=spam_email['email_from'].apply(split_it)
spam_email['Sender'] = spam_email['Sender'].map(lambda x: str(x)[2:-2])
```

```python
df = spam_email.drop(spam_email.columns[[0, 3,4]], axis=1)
df
```
